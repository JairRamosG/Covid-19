```python
import time
import csv
```

```python
nombreArch = "DatosCovid/220720COVID19MEXICO.csv" 
print("IDENTIFICANDO LOS 26 REGISTROS PROBLEMÁTICOS")

ini = time.time_ns()
contador = 1
erroneos = 0
registros_problematicos = []

with open(nombreArch, encoding="latin1", newline='') as File:  
    archivo = csv.reader(File)
    for registro in archivo:
        if contador == 1:  # Saltar encabezado
            contador += 1
            continue
            
        if len(registro) != 40:
            erroneos += 1
            registros_problematicos.append(contador)
            print(f"{erroneos}. Línea {contador}: {len(registro)} columnas")

        contador += 1

fin = time.time_ns()

print("\n" + "="*60)
print("RESUMEN FINAL")
print("="*60)
print(f"Duración: {(fin-ini)/1e9} segundos")
print(f"Total de registros: {contador}")
print(f"Registros problemáticos: {erroneos}")

print("\nLISTA DE LOS 26 REGISTROS A CONSIDERAR:")
print("="*60)
print("borrar_o_corregir = [")
for i, num in enumerate(registros_problematicos):
    print(f"    {num},")
print("]")

# También mostrar agrupados
print("\nAgrupados para mejor visualización:")
print("borrar_o_corregir = [")
grupo = []
for i, num in enumerate(registros_problematicos):
    grupo.append(num)
    if (i + 1) % 5 == 0 or i == len(registros_problematicos) - 1:
        print("    " + ", ".join(str(x) for x in grupo) + ",")
        grupo = []
print("]")
```

```python
def eliminar_registros_problematicos(archivo_entrada, archivo_salida, registros_a_eliminar):    
    print(f"Iniciando limpieza del archivo: {archivo_entrada}")
    print(f"Registros a eliminar: {len(registros_a_eliminar)}")
    print("-" * 60)
    
    inicio = time.time_ns()
    contador = 0
    eliminados = 0
    registros_guardados = 0
    
    # Convertir a set para búsqueda más rápida
    lineas_a_eliminar = set(registros_a_eliminar)
    
    try:
        with open(archivo_entrada, 'r', encoding='latin1', newline='') as entrada, \
             open(archivo_salida, 'w', encoding='UTF-8', newline='') as salida:
            
            lector = csv.reader(entrada)
            escritor = csv.writer(salida)
            
            for registro in lector:
                contador += 1
                
                # Verificar si debemos eliminar esta línea
                if contador in lineas_a_eliminar:
                    eliminados += 1
                    print(f"Línea {contador} eliminada")
                    
                    # Verificar si era un registro problemático (no 40 columnas)
                    if len(registro) != 40:
                        print(f"   (Tenía {len(registro)} columnas en lugar de 40)")
                    
                    continue  # Saltar este registro
                
                # Escribir el registro en el archivo limpio
                escritor.writerow(registro)
                registros_guardados += 1
                
                # Mostrar progreso cada 500,000 registros
                if contador % 500000 == 0:
                    print(f"Procesados: {contador:,} registros...")
        
        fin = time.time_ns()
        tiempo_total = (fin - inicio) / 1e9
        
        print("\n" + "=" * 60)
        print("PROCESO DE LIMPIEZA COMPLETADO")
        print("=" * 60)
        print(f"Tiempo total: {tiempo_total:.2f} segundos")
        print(f"Registros procesados: {contador:,}")
        print(f"Registros eliminados: {eliminados:,}")
        print(f"Registros guardados: {registros_guardados:,}")
        print(f"Archivo limpio guardado en: {archivo_salida}")
        
        # Verificar que todos fueron eliminados
        eliminados_no_encontrados = [x for x in registros_a_eliminar if x > contador]
        if eliminados_no_encontrados:
            print(f"\nAdvertencia: {len(eliminados_no_encontrados)} registros no se encontraron")
            print("   (posiblemente números de línea fuera de rango):")
            for num in eliminados_no_encontrados[:10]:  # Mostrar solo los primeros 10
                print(f"   - Línea {num}")
            if len(eliminados_no_encontrados) > 10:
                print(f"   ... y {len(eliminados_no_encontrados) - 10} más")
        
        return True
        
    except FileNotFoundError:
        print(f"ERROR: No se encontró el archivo '{archivo_entrada}'")
        return False
    except Exception as e:
        print(f"ERROR: {e}")
        return False
    
eliminar_registros_problematicos("DatosCovid/220720COVID19MEXICO.csv", 'DatosCovid/Covid_limpio.csv', registros_problematicos)
```
